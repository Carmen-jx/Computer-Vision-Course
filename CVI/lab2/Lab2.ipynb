1. Read an image, convert to rgb and then to HSV and the show the images.

```python
#import libraries
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./money.JPG") #read the image

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert original image to rgb
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV) #convert to hsv from rgb image

#display the images
cv2.imshow('image', img_rgb)
cv2.imshow('image', img_hsv)
cv2.waitKey(0) 
cv2.destroyAllWindows()

```

4. 
- Add padding to all four sides of the image using a border of 50 pixels.
Try at least three border types. What visual differences do you see?
- How can you change the color of padding

```python
img = cv2.imread("./money.JPG") #read the image

#creating the borders
border1 = cv2.copyMakeBorder(img, 50, 50, 50, 50, cv2.BORDER_CONSTANT, value=[0, 0, 0])
border2 = cv2.copyMakeBorder(img, 50, 50, 50, 50, cv2.BORDER_REFLECT)
border3 = cv2.copyMakeBorder(img, 50, 50, 50, 50, cv2.BORDER_WRAP)

cv2.imshow('image1', border1)
cv2.imshow('image2', border2)
cv2.imshow('image3', border3)

cv2.waitKey(0)
cv2.destroyAllWindows()

#border 1 displays a black border around the image
#border 2 display a mirror reflection up tp 50 pixels of all four sides of the image 
#border 3 wraps around the opposite edge
```

5. You are building a simple image processing pipeline. Your goal is to:

- Keep the original image unchanged.
- Create a version with a red square in the top-left corner.

What would you do?

```python
import copy #import copy library
img = cv2.imread("./money.JPG") #read the original image

#create a deep copy of original image
img_copy = copy.deepcopy(img)

#draw a rectange onto the copy image using the cv2 built in function rectange
cv2.rectangle(img_copy, (0,0), (50,50), (0,0,255), thickness=-1)

#display the copy image with the square
cv2.imshow('image with square', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

6. Plot the function y = x^2+6 using Matplotlib.

```python
x = [i for i in range(1,7)] #initialize x values
y = [(i ** 2 + 6 )for i in x] #initialize y values using the formula

plt.plot(x , y, label = "y=x^2+6") #plots the x and y coordinates
plt.legend()
plt.show() #display the graph
```

7. Plot sin(x) and cos(x) on the same figure. Add grid, labels, title, and legend.

```python
x = np.arange(0, 5, 0.1) #initialize x coordinates from range 0 to 5 with 0.1 diferences
y1 = np.sin(x) #initize sin(x) y-axis values
y2 = np.cos(x) #initialize cos(x) y-axis values

#plot the two coordinate sets
plt.plot(x, y1, label = "sin(x)")
plt.plot(x, y2, label = "cos(x)" )
plt.title("sin(x) and cos(x)") #title for the graph
plt.legend() #shows the legends of the two formula
plt.show() #display the figure
```

8. Read an image and display it alongside its grayscaled version using matplotlib.

```python
plt.figure(figsize=(10,5))

img = cv2.imread("./money.JPG") #read the image

#plot original image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert image to rbg scale
plt.subplot(121)
plt.imshow(img_rgb)
plt.title("Original Image")

#plot grayscaled
plt.subplot(122)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), cmap='gray') #convert image to gray scale and display it
plt.title('Grayscaled Version')

#display the images side by side
plt.show()
```

9. Having an image, create a 1x3 subplot that shows:

- Original RGB image
- Red channel
- HSV version
- Adjust the figure size for better visualization.

```python
plt.figure(figsize=(10,3))

img = cv2.imread("./money.JPG") #read the image


#plot original image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert image to rbg scale
plt.subplot(131)
plt.imshow(img_rgb)
plt.title("Original RGB Image")

#plot original
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV) #convert to hsv from rgb image
plt.subplot(132)
plt.imshow(img_hsv)
plt.title("HSV Image")

#plot red channel
img_red = img[:, :, 2]
plt.subplot(133)
plt.imshow(img_red, cmap='Reds')
plt.title("Red Channel")

#display all 3 
plt.show()

```
