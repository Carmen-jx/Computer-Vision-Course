```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
x = [1, 2, 3, 4, 5, 6] #x = [1 for i in range(1, 7)]
y = [i ** 2 for i in x]

#plt.figure(figsize=(1,4))
plt.plot(x,y)
plt.title("Math function: y = x ** 2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.savefig("plot.png") #save the plot
plt.show() #need to wriote aft er every plot for good practice
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2, markersize=8)
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y)
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i for i in x]

y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x^2")
plt.plot(x, y3, label="y = x^3")

plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
#shallow copy vs. Deep copy
import copy

a = [4,5,6]
b =a
print(f'a: {a}, b: {b}')
a[0] = 0
print(f'a: {a}, b: {b}') #both a and b are changed, this is shallow copy

a = [4,5,6]
b = a.copy() #deep copy, becareful if this is nested list ex [4, [5,5], 6]
#b = copy .deepcopy(a) #deep copy, both works 
print(f'a: {a}, b: {b}')
a[0] = 0
print(f'a: {a}, b: {b}') #only a is changed, b is not changed
```

```python
img1 = cv2.imread("plot.png")
img2 = cv2.imread("plot.png") 

print(type(img1)), type(img2)

img1_copy = img1
img2_copy = img2.copy()

img1[0:100, 0:100] = 0
img2[100:200, 100:200] = 0

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("img1_copy", img1_copy)
cv2.imshow("img2_copy", img2_copy)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
