##### 1. Search about categorical values and preprocessing steps required to make them compatible with Sklearn.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Green', 'Blue'],
    'Fruit': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Price': [1.2, 0.8, 1.3, 1.0, 0.9, 1.1]
})

df
```

##### 2. Convert the 'Fruit' column into numerical values using label encoding.
Use sklearn's LabelEncoder or pandas .map().

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# convert fruit into numerical values using labelEncoder
le = LabelEncoder()
df['Fruit_Label'] = le.fit_transform(df['Fruit'])

df[['Fruit', 'Fruit_Label']]
```

##### 3. Apply one-hot encoding to the 'Color' column using pandas' get_dummies.
Ensure the result includes all dummy columns.

```python
# use dummies to apply one-hot encoding to 'Color'
color_dummies = pd.get_dummies(df['Color'], prefix='Color')
# concatenate with original DataFrame
df_onehot_color = pd.concat([df, color_dummies], axis=1)

df_onehot_color
```

##### 4. Apply one-hot encoding to the 'Fruit' column using sklearn's OneHotEncoder.
Output the result as a DataFrame.

```python
# apply one-hot encoding to 'Fruit' using OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
fruit_encoded = ohe.fit_transform(df[['Fruit']])

# create DataFrame with column names
fruit_ohe_df = pd.DataFrame(fruit_encoded, columns=ohe.get_feature_names_out(['Fruit']))

#concatenate with original df
df_fruit_onehot = pd.concat([df, fruit_ohe_df], axis=1)

df_fruit_onehot
```

##### 5. Apply one-hot encoding to the 'Color' column but drop the first category to avoid multicollinearity.

```python
# apply one-hot encoding to 'Color' with drop_first=True
# this will drop the first category to avoid multicollinearity
color_dummies_dropped = pd.get_dummies(df['Color'], prefix='Color', drop_first=True)

# concatenate with original DataFrame
# this will result in one less column for 'Color'
df_color_dropped = pd.concat([df, color_dummies_dropped], axis=1)
 
df_color_dropped
```

##### 6. Add an 'Education' column with values ['High School', 'Bachelors', 'Masters', 'PhD']
Then encode it using ordinal encoding in this order: High School < Bachelors < Masters < PhD

df['Education'] = ['Bachelors', 'High School', 'Masters', 'PhD', 'Masters', 'High School']

```python
# create a DataFrame with ordinal categorical columns
df['Education'] = ['Bachelors', 'High School', 'Masters', 'PhD', 'Masters', 'High School']

# create an ordinal mapping for the 'Education' column
education_order = {'High School': 0, 'Bachelors': 1, 'Masters': 2, 'PhD': 3}
df['Education_Ordinal'] = df['Education'].map(education_order)
```
